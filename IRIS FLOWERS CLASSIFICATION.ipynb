```python
# Seperating input column and the output column

data=dataframe.values

X=data[:,0:4]
Y=data[:,4]

print(X)
```

```python
# splitting data for training and testing

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
# support vector machine algorithm

from sklearn.svm import SVC

#model fitting

model_svc=SVC()
model_svc.fit(X_train,Y_train)

predictionSVC = model_svc.predict(X_test)

# Calculating the accuracy

from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,predictionSVC))
```

```python
for i in range(len(predictionSVC)):
    print(Y_test[i],predictionSVC[i])
```

```python
# LOGISTIC REGRESSION

from sklearn.linear_model import LogisticRegression

#model fitting

model_LR=LogisticRegression()
model_LR.fit(X_train,Y_train)

predictionLR = model_LR.predict(X_test)

# Calculating the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,predictionLR))
```

```python
for i in range(len(predictionLR)):
    print(Y_test[i],predictionLR[i])
```

```python
# K-NEAREST NEIGHBORS

from sklearn.neighbors import KNeighborsClassifier

#model fitting

model_KNN=KNeighborsClassifier()
model_KNN.fit(X_train,Y_train)

predictionKNN = model_KNN.predict(X_test)

# Calculating the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,predictionKNN))
```

```python
for i in range(len(predictionKNN)):
    print(Y_test[i],predictionKNN[i])
```

```python
# DECISION TREE CLASSIFIER

from sklearn.tree import DecisionTreeClassifier

#model fitting

model_DTC=DecisionTreeClassifier()
model_DTC.fit(X_train,Y_train)

predictionDTC = model_DTC.predict(X_test)

# Calculating the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,predictionDTC))
```

```python
for i in range(len(predictionDTC)):
    print(Y_test[i],predictionDTC[i])
```

```python
# A DETAILED CLASSIFICATION REPORT

from sklearn.metrics import classification_report
print(classification_report(Y_test,predictionDTC))
```

```python
# PREDICTION OF THE SPECIES FROM THE INPUT VECTOR

X_new=np.array([[3,2,1,0.2],[4.9,2.2,3.8,1.1],[5.3,2.5,4.6,1.9]])

prediction= model_DTC.predict(X_new)
print("Prediction of species: {}", format(prediction))
```
